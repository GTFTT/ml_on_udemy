```python
import matplotlib.pyplot as plt
```

```python
# Let's make gradient descent(find minimum value) of a function: f(x) = x^2 + x + 1
# Its derivative is: f'(x) = 2x + 1

# alpha = 1
alpha = 0.001
loss = []
random_starting_point = 100
curr_point = random_starting_point
epochs = 20000

for i in range(epochs):
    curr_point = curr_point - alpha * (2 * curr_point + 1)
    loss.append(curr_point ** 2 + curr_point + 1)
```

```python
plt.plot(loss)
print('Min loss: ', min(loss))
print('Current point:', curr_point)
```
